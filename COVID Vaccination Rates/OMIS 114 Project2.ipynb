#### Cleaned Employment Dataframe

```python
new_header = dfe.iloc[0] #Changes the first row to be the header of the dataframe
dfe = dfe[1:]
dfe.columns = new_header
```

```python
#Counts how many columns in the table there are and drops all columns past "age over 75"
h = 0
for i in dfe.columns:
    if 'Margin of Error' in i:
        dfe.drop(columns = i, inplace = True)
    elif h > 12:
        dfe.drop(columns = i, inplace = True)
    else:
        h += 1
```

```python
#Removes the id column
dfe.drop(columns = 'id', inplace = True)
```

```python
#Renames the columns into simpler terms
dfe.columns = ['Zip Code',
              'Total Population Age 16+',
              'Ages 16-19',
              'Ages 20-24',
              'Ages 25-29',
              'Ages 30-34',
              'Ages 35-44',
              'Ages 45-54',
              'Ages 55-59',
              'Ages 60-64',
              'Ages 65-74',
              'Ages 75+']
```

```python
#Defines a function to remove ZTCA5 from every zipcode
def Zipfinder(Zipcode):
        j = Zipcode.replace('ZCTA5 ', '')
        return j
```

```python
dfe['Zip Code'] = dfe['Zip Code'].apply(Zipfinder) #Applies Zipfinder to the Zipcode column
```

```python
#Converts columns into integers
def Intconvert(integer):
        i = int(integer)
        return i
```

```python
dfe['Total Population Age 16+'] = dfe['Total Population Age 16+'].apply(Intconvert)
```

```python
dfe['Ages 16-19'] = dfe['Ages 16-19'].apply(Intconvert)
```

#### Cleaned Population Dataframe

```python
#Converts Zip Code values into a string
def Strconvert(string):
        z = str(string)
        return z
```

```python
dfp['Zip Code'] = dfp['Zip Code'].apply(Strconvert)
```

```python
dfp.rename(columns = {'Population' : 'Total Population'}, inplace = True)
```

#### Cleaned Vehicle Dataframe

```python
dfv.drop(columns = ['Date', 'Duty', 'Fuel'], inplace = True) #Drops the columns date, duty, and fuel
```

```python
dfv.drop(index = dfv[dfv['Zip Code'] == 'OOS'].index, inplace = True) #Drops all rows with OOS as the zip code value
```

```python
dfv #Vehicle dataframe
```

```python
dfvg = dfv.groupby('Zip Code')[['Make']].max() #Groups all vehicles by zip code and finds the most used make and stores it into dfvg
dfvg
```

#### Cleaned Vaccinations Dataframe

```python
print(dfc.columns)
```

```python
#print(dfc[['as_of_date', 'zip_code_tabulation_area', 'percent_of_population_fully_vaccinated']])

w = dfc[dfc['as_of_date'] == '2022-01-04']
print(w[['zip_code_tabulation_area', 'percent_of_population_fully_vaccinated']])

```

```python
#For our purpouses we only want to look at vaccination rates as of the begining of this year 

dfc.drop(dfc[dfc['as_of_date'] != '2022-01-04'].index, inplace = True)
#dfc
```

```python
# For our purpouses we're only keeping the zip code column and the columns with %vaccinated info

dfc = dfc.drop(['as_of_date', 'local_health_jurisdiction',
       'county', 'vaccine_equity_metric_quartile', 'vem_source',
       'age12_plus_population', 'age5_plus_population',
       'persons_fully_vaccinated', 'persons_partially_vaccinated', 'redacted'], axis=1)

#dfc
```

```python
#we don't want the rows with nan values for % vaccinated 

dfc = dfc.dropna(subset = ['percent_of_population_fully_vaccinated', 'percent_of_population_partially_vaccinated'])
dfc

```

```python
#we'll give the columns easier names 

dfc = dfc.rename(columns = {"zip_code_tabulation_area": "Zip Code", "percent_of_population_fully_vaccinated": "Fully Vaccinated", "percent_of_population_partially_vaccinated": "Partially Vaccinated", "percent_of_population_with_1_plus_dose": "1_plus_dose"})
```

```python
#Converts the Zip Code column into strings
dfc['Zip Code'] = dfc['Zip Code'].apply(Strconvert)
```
