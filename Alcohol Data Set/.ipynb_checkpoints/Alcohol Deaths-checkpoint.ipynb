## Project Outline

Analyze the alcohol specific deaths in the UK and identify leading causes of death

```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"


import sklearn
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
import joblib
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter
```

```python
UK = "Alcohol-specific deaths in the UK.csv"
England = "Alcohol-specific deaths in England.csv"
N_Ireland = "Alcohol-specific deaths in Northern Ireland.csv"
Wales = "Alcohol-specific deaths in Wales.csv"
```

```python
#Clean the UK dataframe
dfuk = pd.read_csv(UK)
dfuk.dropna(inplace=True)
dfuk.rename(columns={"Year [note 3]":"Year"}, inplace=True)
dfuk.drop(labels='ICD-10 code', axis=1, inplace=True)
dfuk = dfuk.astype({'Year' : 'int'}, copy=False)
```

```python
dfuk
```

```python
years = dfuk.groupby('Year')['All ages'].sum().reset_index()
years
```

```python
value_counts = Counter(dfuk['Year'])
X = list(years['Year'])
y = list(years['All ages'])

plt.xticks(np.arange(min(X), max(X)+1, 2.0))
plt.plot(X,y,'bo-')
```

```python
test_dfuk = dfuk.groupby(['Year','Sex'])['Sex'].count()
```

```python
X = np.array(X)
y = np.array(y)

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
X_train.shape
X_test.shape
```

```python
data = pd.DataFrame({'X_train': X_train.reshape(-1), 
                     'y_train': y_train})
data.head()
```

```python
import statsmodels.formula.api as smf
model = smf.ols('y_train ~ X_train', data = data)
model = model.fit()
print(model.summary())
```

```python
# Predict values for the train set
y_train_pred = model.predict()
y_train_pred

# Predict values for the test set
y_test_pred = model.predict({'X_train': X_test})
y_test_pred
```

```python
#Plot both graphs
plt.plot(X_train, y_train_pred) # regression line
plt.plot(X_train, y_train, 'ro') # scatter plot showing actual data
plt.title('Actual vs Predicted')
plt.xlabel('Year')
plt.ylabel('Number of Deaths')
plt.xticks(np.arange(min(X), max(X)+1, 2.0))
plt.show()
```
